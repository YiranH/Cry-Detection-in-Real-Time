Generate timestamps from frames of positive data



```
def toTimeStamp(frames):
  secs = frames / 24.0
  m = secs // 60
  s = secs % 60
  return int(m), round(s, 2)

def getDuration(time):
  dur = time / 24.0
  return round(dur, 2)
```

```
PATH = "/content/drive/My Drive/636/data/raw_txt/"
FILENAME = "cry.txt"
filepath = PATH + FILENAME
```

```
!rm -rf '00.txt' 
outfile = open('00.txt', 'wt')

with open(filepath, "r") as f:
  lines = f.readlines()
  for line in lines:
    p_frame = line.replace('\n', '').split(' ')
    p0 = int(p_frame[0])
    p1 = int(p_frame[1])
    start_m, start_s = toTimeStamp(p0)
    duration = getDuration(p1 - p0)
    outfile.write('00 %02d %.2f %.2f\n'% (start_m, start_s, duration))
outfile.close()
```

```
OUTFILENAME = '00.txt'
uploaded = drive.CreateFile({'title': OUTFILENAME})
uploaded.SetContentFile(OUTFILENAME)
uploaded.Upload()
print('Uploaded file with ID {}'.format(uploaded.get('id')))
```

Generate timestamps of negative data

```
def getDur(m0, s0, m1, s1):
  t0 = m0 * 60 + s0
  t1 = m1 * 60 + s1
  return t1 - t0
```

```
def getTS(filepath, outname):
  outfile = open(outname, 'wt')
  with open(filepath, "r") as f:
    lines = f.readlines()
    for line in lines:
      times = line.replace('\n', '').replace(':', ' ').split(' ')
      m0 = int(times[0])
      s0 = int(times[1])
      m1 = int(times[2])
      s1 = int(times[3])
      duration = getDur(m0, s0, m1, s1)
      outfile.write('00 %02d %.2f %02d\n'% (m0, s0, duration))
  outfile.close()
```

```
for i in range(1, 9):
  filename = '0' + str(i) + '.txt'
  filepath = PATH + filename
  outname = filename
  getTS(filepath, outname)
  uploaded = drive.CreateFile({'title': outname})
  uploaded.SetContentFile(outname)
  uploaded.Upload()
  print('Uploaded file with ID {}'.format(uploaded.get('id')))
```
